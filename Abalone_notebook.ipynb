```python
# Import Dataset
train_df = pd.read_csv("./Data/train.csv")
test_df = pd.read_csv("./Data/test.csv")
```

## Data Preprocessing

```python
# Import Libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_df.drop(["Rings"], axis = 1), train_df["Rings"], test_size = 0.2, random_state = 42)

X_train.reset_index(drop = True, inplace = True)
X_val.reset_index(drop = True, inplace = True)
y_train.reset_index(drop = True, inplace = True)
y_val.reset_index(drop = True, inplace = True)
```

```python
display(X_train)
display(X_val)
```

```python
ohe = OneHotEncoder(sparse_output = False)
ohe.fit(X_train[["Sex"]])

# OHE X_train
train_sex_ohe = ohe.transform(X_train[["Sex"]])
train_sex_ohe_df = pd.DataFrame(train_sex_ohe, columns = ohe.get_feature_names_out())
X_train = pd.concat([X_train, train_sex_ohe_df], axis = 1)

# OHE X_val
val_sex_ohe = ohe.transform(X_val[["Sex"]])
val_sex_ohe_df = pd.DataFrame(val_sex_ohe, columns = ohe.get_feature_names_out())
X_val = pd.concat([X_val, val_sex_ohe_df], axis = 1)

# OHE test_df
test_sex_ohe = ohe.transform(test_df[["Sex"]])
test_sex_ohe_df = pd.DataFrame(test_sex_ohe, columns = ohe.get_feature_names_out())
test_df = pd.concat([test_df, test_sex_ohe_df], axis = 1)
```

```python
# Drop columns that will not be useful for prediction task
X_train.drop(["id", "Sex"], axis = 1, inplace = True)
X_val.drop(["id", "Sex"], axis = 1, inplace = True)
X_test = test_df.drop(["id", "Sex"], axis = 1)
```

## Model Development

```python
# Import Libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf_reg = RandomForestRegressor(random_state = 42)
```

```python
# Default (Random Forest Regressor) model performance
rf_reg.fit(X_train, y_train)

train_set_r2 = rf_reg.score(X_train, y_train)
val_set_r2 = rf_reg.score(X_val, y_val)

print(f"R^2 (Train): {train_set_r2}")
print(f"R^2 (val): {val_set_r2}")
```

<p> Results indicate possible overfitting of model. Proceed to feature selection and model/hyperparameter tuning. </p>

```python
rf_reg.feature_importances_
```

```python
# Feature importance
feature_importances = rf_reg.feature_importances_

# Sort the feature importance scores 
sorted_indices = np.argsort(feature_importances)
sorted_feature_importances = feature_importances[sorted_indices]
sorted_features = [X_train.columns[i] for i in sorted_indices]

plt.figure(figsize = (15, 10))
plt.barh(sorted_features, sorted_feature_importances)
plt.title("Feature Importance of Abalone on Rings")
```

```python
corr = X_train.corr()

plt.figure(figsize = (15, 10))
sns.heatmap(corr, annot = True)
```

```python
# Explore a model with reduced features
potential_reduced_features = ["Shell weight", "Whole weight.1", "Length", "Height", "Diameter"]

# Declare model and train on reduced set of features
rf_reg_reduced_feat = RandomForestRegressor(random_state = 42)
rf_reg_reduced_feat.fit(X_train[potential_reduced_features], y_train)

# Print performance on train and test set
print(f'R^2 (train [reduced features]) {rf_reg_reduced_feat.score(X_train[potential_reduced_features], y_train)}')
print(f'R^2 (val [reduced features]) {rf_reg_reduced_feat.score(X_val[potential_reduced_features], y_val)}')
```

<p> Reduced feature set does not improve performance on both train and validation set. </p>
<p> Hence, will continue with the default/original set of features for the model. </p?>
